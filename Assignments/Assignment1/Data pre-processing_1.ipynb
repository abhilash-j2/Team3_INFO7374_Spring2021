```python
def sanitize_names(string_list):
    string_list = [strg.strip() for strg in string_list]
    string_list = [re.sub('\s','_', strg) for strg in string_list]
    string_list = [re.sub('[^A-Za-z0-9]','_', strg) for strg in string_list]
    return string_list
```

```python
pref_files = glob.glob("data/raw/ADS16_Benchmark_part?/Corpus/Corpus/U*/U*-PREF.csv")
pref_files
```

```python
def collate(filepaths):
    df_list = [read_file_add_id(filepath) for filepath in filepaths]
    df_list = pd.concat(df_list)
    df_list.columns = sanitize_names(df_list.columns)
    df_list = df_list.reset_index(drop=True)
    return df_list
```

# Collate Rating

```python
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
rate_filepaths = glob.glob(r"D:\data\ADM\Assignment_1\ADS16_Benchmark_part*\ADS16_Benchmark_part*\Corpus\Corpus\U*\U*-RT.csv")
```

```python
def read_file_add_id_rating(filepath):
    user_id = get_userid(filepath)
    df = pd.read_csv(filepath,sep=';',header=1)
    #f['used_id'] = user_id
    return df
```

```python
rate_dfs = [read_file_add_id_rating(filepath) for filepath in rate_filepaths]
```

```python
rate = pd.concat(rate_dfs)
```

```python
rate.reset_index(drop=True,inplace=True)
rate
```

```python
j=0
while j <120:
    sample = rate_dfs[j]

    array = []
    ary = []

    for i in sample.columns:
        array.append(sample[i][0])

    for i in zip(sample.columns,array):
        ary.append(i)

    frame = pd.DataFrame(ary)

    df = frame[1].str.split(',',expand=True)

    frame.drop(1,axis=1,inplace=True)

    frame.rename(columns={0:'category'},inplace=True)


    frame1=pd.concat([frame,df],axis=1)

    frame1 = frame1.melt(id_vars=['category'])


    frame1.sort_values(by=['category','variable'],inplace=True)


    frame1.reset_index(drop=True,inplace=True)

    index=pd.MultiIndex.from_frame(frame1[['category','variable','value']])

    frm = pd.DataFrame(index =index,columns=['Value'])

    frm.reset_index(inplace=True, level = ['value'])

    frm.drop(['Value'],axis=1,inplace=True)

    pd.set_option('display.max_columns',None)

    frm = frm.transpose()
    
    if j == 0:
        frm1 = frm
    if j >0:
        frm1 = pd.concat([frm1,frm])
    j +=1
```
